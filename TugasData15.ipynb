```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('upah_minimum_regional_kabupaten_sukabumi.csv')
df
```

```python
plt.figure(figsize=(10,5))
plt.pie(df['umr'], labels=df['tahun'], startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
